## Total Fares by City Type

```python
# Calculate Type Percents

total_fare = ride_data["fare"].sum()

group_type = merged.groupby("type")
fare_cost = group_type["fare"].sum()
total_fare_city = round((total_fare.sum()),2)
percent = fare_cost/total_fare_city * 100

print(percent)
# Build Pie Chart

# Save Figure
```

```python
# Show Figure

labels = ["Rural", "Suburban", "Urban"]
sizes = [6.8, 30.5, 62.7]
colors = ["gold", "lightblue", "lightcoral"]
explode = (0, 0, 0.15)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title("% of Total Fares by City Type")

plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

total_rides = ride_data["ride_id"].sum()

group_type = merged.groupby("type")
ride_count = group_type["ride_id"].sum()
total_ride_count = round((ride_count.sum()),2)
percent = ride_count/total_ride_count * 100

print(percent)

# Build Pie Chart

# Save Figure
```

```python
# Show Figure

labels = ["Rural", "Suburban", "Urban"]
sizes = [5, 26.8, 68.2]
colors = ["gold", "lightblue", "lightcoral"]
explode = (0, 0, 0.15)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title("% of Total Rides by City Type")

plt.show()
```
